2 - Tente atrair a atenção do público.

3 - Use adequadamente paletas pré-definidas.

4 - Observe sempre o contexto e objetivo da visualização.

5 - Considere as emoções transmitidas pela cor com tranquilidade, entusiasmos, seriedade e muito mais.

6 - Considere o tipo de dado: categóricos, sequênciais, divergentes para escolha da paleta.

7 - Evite cores muito brilhantes e saturadas, pois podem ser muito intensas e distrair a atenção do público.

8 - Pedir feedback para outras pessoas.

9 - Salvar o gráfico para o público-alvo.

10 - Se possível, aplicar o storytelling.

```python
top_10 = df.sort_values(by='Total', ascending=False).head(10)
top_10
```

```python
sns.set_theme()
```

```python
ax = sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h')
ax.set(
    title="Os dez Países com mais imigrantes para o Canadá",
    xlabel="Total de Imigrantes",
    ylabel=""
)
plt.show()
```

```python
def gerar_grafico_paleta(pallete):
  fig,ax = plt.subplots(figsize=(8,4))
  sns.barplot(data=top_10,y=top_10.index,x='Total',orient='h',palette=pallete)
  ax.set_title("Os dez Países com mais imigrantes para o Canadá",
      loc='left',
      fontsize=16)

  ax.set_xlabel("Total de Imigrantes",fontsize=14)
  ax.set_ylabel("")

  plt.show()

```

```python
gerar_grafico_paleta('rainbow_r')
```
